```python
# 열(columns) -> 행(row) : DF['col명']['idx명'] , DF.col명['idx']
# 행(row) -> 열(columns) : DF.loc['idx',col명],  DF.loc['idx']['col']
#                        : DF.iloc[idx번호,col번호]

print(df['영어']['인아']) # series data 를 먼저 뽑고, 그 다음 idex를 언급
print(df.영어['인아'])
print(df.loc['진수']['영어'])
print(df.loc['진수','영어'])
print(df.iloc[3])
print(df.iloc[3,1])
```

```python
df

#DF.loc['idx',col명]            -> 단일 원소
#DF.loc[ [,,],  [,,] ]          -> 불규칙적인 순서를 가진 다수 원소들 조회
df.loc[['서준','진희','인아']]
df.loc[['서준','진희','인아'],['수학','영어']]
```

```python
# DF.iloc[ [] , [] ]
df.iloc[[0,-1]]  # df.iloc[0,-1]과는 차이가 있다.
# df.iloc[idx, col]      : 단일 원소
# df.iloc[[,,] , [,,]]   : 다수 원소
df.iloc[[0,-1],[0,-1]]
```

```python
data1 = {'kor':100,'eng':80, 'mat':90}
data2 = {'eng': 85,'mat':95, 'phy':89}
sr1 = pd.Series(data1)
sr2 = pd.Series(data2)
print(sr1)
print()
print(sr2)
#Series + 연산자(+-*/) + 숫자 : 나열된 순서와 상관없이 동일 indx 명끼리
                                연산이 이루어지고, 서로 없는 index 끼리는 연산이 이루어지지 않고 NaN으로 대체
```

```python
sr1 + sr2
```

```python
# * 사전형에서 NaN을 방지하기 위해서 a.get 형식으로 하면 된다. A.get('phy',0) -> 없는 값은 0으로 출력
#Series.연산메서드(Series) :NaN(결측치)에 대한 대체값으로 연산도 가능하다.
#add(+),sub(-),mul(*),div(/)
print(sr1+ sr2)
print()
print(sr1.add(sr2))
print()
print(sr1.add(sr2,fill_value = 0))
```
