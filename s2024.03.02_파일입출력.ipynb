```python
import pandas as pd
```

```python
#pandas.read_csv('csv파일',옵션) : csv 파일을 읽어들여 DataFrame으로 생성한다.
#encoding = 'utf-8','ansi','euckr' : 문자가 저장된 인코딩 방식을 지정한다.
#separator = '구분자'              : 값들을 구분하는 문자 지정한다. 기본값은 ','
#header, index_col, skiprows
#header = 'index행'                : 컬럼명을 갖고 있는 행 index 번호, 기본값 : 0

#encoding = 'utf-8', header = 0
filename = 'csv/exam1.csv'
df = pd.read_csv(filename, header = 0)
df
#utf-8방식이 젤 잘불러와짐, 여기서는 옵션도 안줬음.
df.set_index('name',inplace=True)
df
```

```python
filename = 'csv/exam2.csv'

#인코딩방식이 'utf-8'이 아닌 경우에는 encoding = '방식', ansi와 euckr은 호환이 된다.
df = pd.read_csv(filename, header = 0, encoding = 'ansi')
#index_col = 'col명'   : DF 생성시 특정 column을 index(축)ㅇ로 지정한다.
#                        set_index('col명')을 생략할 수 있다.
df = pd.read_csv(filename, header = 0, encoding = 'ansi', index_col = 'name')
df
```

```python
filename = 'csv/exam3.csv'

#header = None : 원래 파일에 columns(필드)명을 포함하지 않는 경우 생략할 수 있다.
#names = [col배열] : DataFrame 객체 생성시 별도의 columns명들을 지정할 수 있다.

#기존 csv 파일에는 columns명으로 사용할 행이 없으므로header = None 
#그리고, 사용자가 임의로 columns명들을 지정 name = [columns배열]
#그중 columns(필드)들 중 '이름' 컬럼을 index(축)으로 지정했다.
df = pd.read_csv(filename, header = None, encoding = 'euckr',
                 names = ['이름','주소','국어','영어','수학','과학','역사'],
                 index_col = '이름')
df
```

```python
filename = 'csv/exam4.csv'

#skiprows = 행수 : csv 파일을 읽어들일 때 최초로 무시하고 지나칠 행수, 공백행은 없는 걸로 친다.

df = pd.read_csv(filename,encoding='euckr',skiprows=2)
df
```

```python
filename = 'csv/exam5.tsv'

#tab은 \t로 보임
#sep \ '구분자'  : 값(value)들 사이의 구분자를 지정, 기본값 ','
# -csv (Comma Separated Values)
# -tsv (Tab Separated Values)
df = pd.read_csv(filename, sep = '\t')
df
```

```python
scores = [80,90,87,78]

sum(scores) / len(scores)
```

```python
total = 0
for i in scores:
    total = i
avg = total / len(scores)
avg
```

```python
def average(scores):
    avg = sum(scores) / len(scores)
    return avg
```

```python
def grade(avg):
    
    if avg >= 90:
        grd = "A"
    elif avg >= 80:
        grd = "B"
    elif avg >= 70:
        grd = "C"
    elif avg >= 60:
        grd = "D"
    else:
        grd = "F"
    return grd
```

```python
for n in range(1,4):
    df = pd.read_excel('excel/students.xlsx',sheet_name = f'{n}반', index_col = 'name')
    df['avg'] = df.loc[:,'kor':].apply(average,axis='columns')
    df['grade'] = df['avg'].apply(grade)
    df.to_excel(f'excel/{n}반_결과.xlsx')
```
