```python
import pandas as pd
```

```python
#pandas.read_csv('csv파일',옵션) : csv 파일을 읽어들여 DataFrame으로 생성한다.
#encoding = 'utf-8','ansi','euckr' : 문자가 저장된 인코딩 방식을 지정한다.
#separator = '구분자'              : 값들을 구분하는 문자 지정한다. 기본값은 ','
#header, index_col, skiprows
#header = 'index행'                : 컬럼명을 갖고 있는 행 index 번호, 기본값 : 0

#encoding = 'utf-8', header = 0
filename = 'csv/exam1.csv'
df = pd.read_csv(filename, header = 0)
df
#utf-8방식이 젤 잘불러와짐, 여기서는 옵션도 안줬음.
df.set_index('name',inplace=True)
df
```

```python
filename = 'csv/exam2.csv'

#인코딩방식이 'utf-8'이 아닌 경우에는 encoding = '방식', ansi와 euckr은 호환이 된다.
df = pd.read_csv(filename, header = 0, encoding = 'ansi')
#index_col = 'col명'   : DF 생성시 특정 column을 index(축)ㅇ로 지정한다.
#                        set_index('col명')을 생략할 수 있다.
df = pd.read_csv(filename, header = 0, encoding = 'ansi', index_col = 'name')
df
```

```python
filename = 'csv/exam3.csv'

#header = None : 원래 파일에 columns(필드)명을 포함하지 않는 경우 생략할 수 있다.
#names = [col배열] : DataFrame 객체 생성시 별도의 columns명들을 지정할 수 있다.

#기존 csv 파일에는 columns명으로 사용할 행이 없으므로header = None 
#그리고, 사용자가 임의로 columns명들을 지정 name = [columns배열]
#그중 columns(필드)들 중 '이름' 컬럼을 index(축)으로 지정했다.
df = pd.read_csv(filename, header = None, encoding = 'euckr',
                 names = ['이름','주소','국어','영어','수학','과학','역사'],
                 index_col = '이름')
df
```

```python
filename = 'csv/exam4.csv'

#skiprows = 행수 : csv 파일을 읽어들일 때 최초로 무시하고 지나칠 행수, 공백행은 없는 걸로 친다.

df = pd.read_csv(filename,encoding='euckr',skiprows=2)
df
```

```python
scores = [80,90,87,78]

sum(scores) / len(scores)
```

```python
total = 0
for i in scores:
    total = i
avg = total / len(scores)
avg
```

```python
def average(scores):
    avg = sum(scores) / len(scores)
    return avg
```

```python
def grade(avg):
    
    if avg >= 90:
        grd = "A"
    elif avg >= 80:
        grd = "B"
    elif avg >= 70:
        grd = "C"
    elif avg >= 60:
        grd = "D"
    else:
        grd = "F"
    return grd
```

### 1개의 excel 파일에 여러개의 sheet로 저장하는 방법
### 1. excel의 워크북 객체 생성
변수 = pd.ExcelWriter('파일명')

### 2. 워크북의 객체에 반복적으로 여러개의 sheet를 저장
데이터프레임1.to_excel(워크북객체, sheet_name = '시트명')<br>
데이터프레임2.to_excel(워크북객체, sheet_name = '시트명')<br>
데이터프레임3.to_excel(워크북객체, sheet_name = '시트명')

### 3. 워크북 객체를 종료
워크북객체.close()

```python
# 1. excel의 워크북객체 생성
excelfile = pd.ExcelWriter('excel/students_result2.xlsx')

# 2. 워크북의 객체에 반복적으로 여러개의 sheet를 저장
df1 = pd.read_excel('excel/students.xlsx', sheet_name = '1반',index_col = 'name')
df1['avg'] = df1.loc[:,'kor':].apply(average, axis= 'columns')
df1['grade'] = df1.avg.apply(grade)
df1.to_excel(excelfile,sheet_name = '1반')

df2 = pd.read_excel('excel/students.xlsx', sheet_name = '2반',index_col = 'name')
df2['avg'] = df2.loc[:,'kor':].apply(average, axis= 'columns')
df2['grade'] = df2.avg.apply(grade)
df2.to_excel(excelfile,sheet_name = '2반')

df3 = pd.read_excel('excel/students.xlsx', sheet_name = '3반',index_col = 'name')
df3['avg'] = df3.loc[:,'kor':].apply(average, axis= 'columns')
df3['grade'] = df3.avg.apply(grade)
df3.to_excel(excelfile,sheet_name = '3반')

# 3. 워크북 객체를 종료
excelfile.close()
```

```python
# for 문으로 바꾸기

# 1. excel의 워크북객체 생성
excelfile = pd.ExcelWriter('excel/students_result3.xlsx')

# 2. 워크북의 객체에 반복적으로 여러개의 sheet를 저장
for i in range(1,4):
    df = pd.read_excel('excel/students.xlsx', sheet_name = f'{i}반',index_col = 'name')
    df['avg'] = df.loc[:,'kor':].apply(average, axis= 'columns')
    df['grade'] = df.avg.apply(grade)
    df.to_excel(excelfile,sheet_name = f'{i}반')

# 3. 워크북 객체를 종료
excelfile.close()
```
