```python
df = pd.read_excel('excel/서울지역 대학교 위치.xlsx')
df.head(3)
```

```python
seoul_map = folium.Map(location=[37.55,126.98],zoom_start = 11)
folium.Marker([df.loc[0]['위도'],df.loc[0]['경도']], popup = df.loc[0]['학교명']).add_to(seoul_map)
folium.Marker([df.loc[1]['위도'],df.loc[1]['경도']], popup = df.loc[1]['학교명']).add_to(seoul_map)
folium.Marker([df.loc[2]['위도'],df.loc[2]['경도']], popup = df.loc[2]['학교명']).add_to(seoul_map)
```
