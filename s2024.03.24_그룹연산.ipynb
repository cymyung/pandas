```python
# p267 그룹객체 필터링
grouped = df.groupby('class')
for key, group in grouped:
    print(key, len(group))
    
# 그룹핑객체.filter(조건함수) : 조건에 부합되는 데이터프레임들을 결합하여 1개의 데이터프레임으로 보여준다.
grouped_filter = grouped.filter(lambda df: len(df)>=200)
grouped_filter
```

```python
# 위의 내용을 함수로...
# filter(함수)는 데이터프레임 객체를 전달받아 true/false 로 반환해야 한다.

def over_200(df):
    return len(df) > 200

grouped.filter(over_200)
```

```python
# titanic의 Nan값을 각 그룹의 평균으로 대체
titanic2 = titanic.copy()
titanic2
```

```python
len(titanic2) - titanic2.count() # 누락갯수 확인
```

```python
titanic2.isnull().sum() # 누락갯수 확인
```

```python
#age를 성별, 아이어른 등으로 구분해서 그 평균으로 대체
#who 컬럼 기준으로 그룹하고, 그룹별 age 컬럼의 평균나이를 조회(시리즈)
grouped = titanic2.groupby('who')
print(grouped)
grouped.apply(titanic2['age'].mean()) # 정답은 담 셀에
```

```python
who_mean_age = titanic2.groupby('who')['age'].mean()
who_mean_age
```

```python
#만약 groupby를 모른다면 아래와 같이 해야 함.

mask = titanic2['who'] =='child'
df = titanic2[mask]
df['age'].fillna(who_mean_age['child'])

mask = titanic2['who'] =='man'
df = titanic2[mask]
df['age'].fillna(who_mean_age['man'])

mask = titanic2['who'] =='woman'
df = titanic2[mask]
df['age'].fillna(who_mean_age['woman'])
```

```python
# 이걸 for 문으로
for who in who_mean_age.index:
    print(who)
```

```python
df_list =[]
for who in who_mean_age.index:
    mask = titanic2['who'] == who
    df = titanic2[mask]
    df['age'] = df['age'].fillna(who_mean_age[who])
    df_list.append(df)
dfs = pd.concat(df_list)
dfs.sort_index(inplace = True)
dfs[20:50]
ori_age = titanic2['age']
dfs.insert(4,'ori_age',ori_age)
dfs[25:50]
```
