```python
#점수에 따른 등급을 함수로
def grade(score):
    if   score >= 90: grd = 'A'
    elif score >= 80: grd = 'B'
    elif score >= 70: grd = 'C'
    elif score >= 60: grd = 'D'
    else:             grd = 'F'
    return grd

students = pd.read_csv('csv/class_1.csv',encoding = 'euckr',header=None,
                       names = ['name','addr','kor','eng','math','sci','his'],
                       index_col = 'name')
students['avg'] = students.iloc[:,1:].mean(axis = 'columns')
students['grade'] = students['avg'].apply(grade)
students
```

```python
# 시리즈공급되면 ->시리즈가반환되는 함수

def missing_values(sr):
    return sr.isnull()

# 시리즈 공급되면 -> 값이 반환되는 함수
def missing_values_count(sr):
    return sr.isnull().sum()
```

```python
df= titanic.iloc[15:30][['age','fare']]
df.apply(missing_values,axis = 'index')  #-> 원본데이터와 동일한 데이터프레임의 모양을 만든다.
                                         # index들을 묶어서 컬럼 단위로 함수에 공급
```

```python
df.apply(missing_values_count,axis = 'index')
```

```python
df.apply(missing_values, axis = 'columns')
```

```python
df.apply(missing_values_count, axis = 'columns')
```

```python
print(df.isnull().sum())  # 컬럼별 누락데이터 개수를 알 수 있다.
print()
print(df.isnull().sum().sum())
```

```python
def missing_value(df):
    return df.isnull()

def missing_count(df):
    return df.isnull().sum()

def missing_total_count(df):
    return df.isnull().sum().sum()

df = titanic.loc[15:24,['age','fare','deck']]
```

```python
df.pipe(missing_value)
```

```python
df.pipe(missing_count)
```

```python
df.pipe(missing_total_count)
```
