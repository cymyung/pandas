```python
#위에꺼 함수화
#강사는 아래 내용을 저 위로 옮김
def extract_1day(tr):
    
    td_list = tr.find_all('td')
    del td_list[2]

    date = td_list[0].text.replace('.','-')
    row = [date]
    for td in td_list[1:]:
        value = td.find('span').text.replace(',','')
        row.append(int(value))
    return row
    
```

```python
text = response.text

html = BeautifulSoup(text, 'html.parser')
tr_list = html.find_all('tr',{'onmouseover':'mouseOver(this)'})

rows = []
for tr in tr_list:
    row = extract_1day(tr)
    rows.append(row)
    
stock = pd.DataFrame(rows)
stock
```

```python
stock = pd.DataFrame(rows, columns= ['date','close','open','max','min','volumn'])
stock = stock.set_index('date')
stock = stock.sort_index(ascending = True)
stock.to_csv('csv/삼성전자_주식.csv')
stock
```

```python
#page를 1~5 까지 50일치

code= '005930'
page = 1
url = f'https://finance.naver.com/item/sise_day.naver?code={code}&page={page}'

response = requests.get(url,headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0 x64)'})
text = response.text

html = BeautifulSoup(text, 'html.parser')
html

tr_list = html.find_all('tr',{'onmouseover':'mouseOver(this)'})

rows = []
for tr in tr_list:
    row = extract_1day(tr)
    rows.append(row)
rows
```

```python
code= '005930'

#page = 1
rows = []
for page in range(1,6):
    url = f'https://finance.naver.com/item/sise_day.naver?code={code}&page={page}'

    response = requests.get(url,headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0 x64)'})
    text = response.text

    html = BeautifulSoup(text, 'html.parser')

    tr_list = html.find_all('tr',{'onmouseover':'mouseOver(this)'})

    for tr in tr_list:
        row = extract_1day(tr)
        rows.append(row)
    sleep(1)
rows
```

```python
stock3 = pd.DataFrame(rows, columns= ['날짜','종가','open','max','min','volumn'])
stock3 = stock3.set_index('날짜')
stock3 = stock3.sort_index(ascending = True)
stock3.to_csv('csv/삼성전자_주식3.csv',encoding = 'utf-8-sig')
stock3
```
